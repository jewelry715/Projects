```python
import tensorflow as tf
import numpy as np

learning_rate=0.1
tf.set_random_seed(0)
np.random.seed(0)

x_data=[[0,0],[0,1],[1,0],[1,1]]
y_data=[[0],[1],[1],[0]]

x=tf.placeholder(tf.float32,[None,2])
y=tf.placeholder(tf.float32,[None,1])

W1=tf.Variable(tf.random_normal([2,5]),name='weight1')
b1=tf.Variable(tf.random_normal([5]),name='bias1')
layer1=tf.sigmoid(tf.matmul(x,W1)+b1)

W2=tf.Variable(tf.random_normal([5,1]),name='weight2')
b2=tf.Variable(tf.random_normal([1]),name='bias2')
hypothesis=tf.sigmoid(tf.matmul(layer1,W2)+b2)
```

```python
cost=-tf.reduce_mean(y*tf.log(hypothesis)+(1-y)*tf.log(1-hypothesis))
train=tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

predicted=tf.cast(hypothesis>0.5,dtype=tf.float32)
accuracy=tf.reduce_mean(tf.cast(tf.equal(predicted,y),dtype=tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        sess.run(train,feed_dict={x:x_data,y:y_data})
        if step%100==0:
            print(step,sess.run(cost,feed_dict={x:x_data,y:y_data}),sess.run(W2))
                  
    h,c,a = sess.run([hypothesis,predicted,accuracy],feed_dict={x:x_data,y:y_data})
    print('\nHypothesis: \n',h,'\nCorrect: \n',c,'\Accuracy: \n',a)
```

다른 유닛에 통과하고 시그모이드 통과->정확한 y값을 알지 못해도 0~1 사이의 값을 갖는다고 생각. 0에 가까운 값들이 곱새지게 되며 결국 괸장히 작은 값을 갖게 됨. 결과단에서 깊어질수록 예측하기 힘들다는 결과를 갖게 됨

이 문제를 경사가 사라진다 부르고 아무도 해결하지 못함. 레이어를 지날 때마다 최초값보다 현저하게 작아지므로 값을 전달해도 무의미함

시그모이드 미분하면 최대치가 0.3
활성화 함수로 사용된 시그모이드 특성으로 기울기 소실 문제가 발생.
계속 곱하다 보면 0에 가까워짐. 층을 거쳐갈수록 기울기가 사라져 가중치 수정이 어려워짐.

결국 이 문제를 해결하기 위해 ReLU 함수

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def z(x,w,b):
    return w*x+b

def H(x):
    global w,b
    return sigmoid(w*x+b)
```

```python
w_list=np.linspace(-20,30,51)
x=np.linspace(-5,10,51)
y=np.append(np.zeros(30),np.ones(21))
err=[]
b=0

for w_value in w_list:
    z=w_value*x+b
    err.append(sum((sigmoid(z)-y)**2)/len(y)) #w에 따른 오차제곱 평균
    
print(err)

plt.title('w와 err의 그래프',fontdict=font1)
plt.xlabel('w',fontdict=font2)
plt.ylabel('err',fontdict=font3)
plt.plot(w_list,err)
plt.show()
```

```python
w_list=np.linspace(-7,3,101)
b=0
x=np.linspace(-5,10,51)
y=np.append(np.zeros(30),np.zeros(21))

cost=[]
for w_value in w_list:
    z=w_value*x+b
    h=sigmoid(z)
    c=-y*np.log(h)-(1-y)*np.log(1-h)
    cost.append(sum(c)/len(c))
    
plt.title('기울기 w에 대한 그래프',fontdict=font1)
plt.grid(True)
plt.xlabel('w',fontdict=font2)
plt.ylabel('cost',fontdict=font3)
plt.scatter(w_list,cost)
plt.show()
```
