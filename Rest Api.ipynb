```python
url='http://apis.data.go.kr/B500001/drghtIdexSpiAnals/analsInfoList'
params={
    'hjdCd':'1168058000',
    'stDt':'20190501',
    'edDt':'20190531',
    'numOfRows':10,
    "pageNo":1,
    'serviceKey':'AT77TAGs2Xi5rj11j0aM4p+7LYnbQiXngOa4hJ0xAGxh47QwGUFm/Ix7okTjIXx8nVZ2ykIToySyfg5D/c7YUQ=='
}
```

```python
resp=requests.get(url,params=params)
resp.text
```

```python
# !pip install lxml
soup=BeautifulSoup(resp.text,'lxml') #xml의 subset임
soup
```

```python
header=soup.find('header')
header
```

```python
body=soup.find('body')
body
```

```python
result_code=header.find('resultcode').text
result_msg=header.find('resultmsg').text
result_code, result_msg
```

```python
num_of_rows=body.find('numofrows').text
pageno=body.find('pageno').text
total_count=body.find('totalcount').text
num_of_rows,pageno,total_count
```

```python
items =body.find_all('item')
result_list=[]
for item in items:
    result_list.append({
        'anldt':item.find('anldt').text,
        'anlrst':item.find('anlrst').text,
        "anlval":item.find('anlval').text,
        "dv":item.find('dv').text,
        "hjdcd":item.find('hjdcd').text
    })
```
