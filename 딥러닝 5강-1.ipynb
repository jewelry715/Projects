```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf

np.random.seed(0)
tf.random.set_seed(0)

dataset=np.loadtxt('data/ThoraricSurgery.csv',delimiter=',')
x=dataset[:,:17]
y=dataset[:,17]
```

```python
print(tf.__version__)
```

```python
model=Sequential()
model.add(Dense(30,input_dim=17,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
#learning_rate은 adam 안에 숨어있음.
```

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
#optimizer:훈련과정을 설명함.metrics:훈련을 모니터링하기 위한 지표
model.fit(x,y,epochs=30,batch_size=10)

print('\n Accuracy: %.4f'%(model.evaluate(x,y)[1]))
```

케라스 내부에서는 다른 딥러닝을 사용할 수 있게 지원. 텐서플로를 한눈에 보기 쉽게 정리함

이진분류:binary crossentropy
다중 클래스 분류:categorical cross entropy

최적화:네트워크가 어떻게 업데이트될지 결정

sgd:lr=0.01,rmsprop=lr=0.001 등 adam으로 해결이 됨

evaluate:학습한 모델에 대한 정확도를 평가:테스트데이터/레이블 테스트데이터

[0]:훈련셋,검증셋의 손실,[1]:정확도

predict():첫번째는 예측하고자 하는 데이터

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense

training_data=np.array([[0,0],[0,1],[1,0],[1,1]],'float32')
target_data=np.array([[0],[1],[1],[0]],'float32')

model=Sequential()
model.add(Dense(32,input_dim=2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_accuracy'])

model.fit(training_data,target_data,epochs=1000,verbose=2)

print(model.predict(training_data))
print('\n Accuracy: %.4f'%(model.evaluate(training_data,target_data)[1]))
```
